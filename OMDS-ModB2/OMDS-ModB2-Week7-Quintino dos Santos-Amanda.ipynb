# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```


To first guide the audience's attention toward the last three months, I would make the following changes: 
- I would make the entire line gray with the three month portion of the line being in a high-contrasting color. 
- I would make the line thicker for that three month segment 
- I would potentially add a lightly shaded rectangle in the background of the three month portion to make it stand out even more

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

data = {
    'Concern': [
        'Engine power is less than expected',
        'Tires make excessive noise while driving',
        'Engine makes abnormal/excessive noise',
        'Seat material concerns',
        'Excessive wind noise',
        'Hesitation or delay when shifting',
        'Bluetooth system has poor sound quality',
        'Steering system/wheel has too much play',
        'Bluetooth system is difficult to use',
        'Front seat audio/entertainment/navigation controls'
    ],
    'Concerns per 1,000': [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]
}

df = pd.DataFrame(data)
df = df.sort_values('Concerns per 1,000', ascending=True)

fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.barh(df['Concern'], df['Concerns per 1,000'], color='gray', height=0.7)

for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(left=False, bottom=False, labelbottom=False)

ax.tick_params(axis='y', labelsize=12, colors='gray')

for bar in bars:
    width = bar.get_width()
    ax.text(width - 0.3, bar.get_y() + bar.get_height()/2, 
            f'{width}', 
            ha='right', va='center', color='white', fontsize=11)


fig.text(0.02, 0.95, 'Top 10 design concerns', fontsize=18, color='gray', ha='left')

ax.text(1.0, 1.02, 'concerns per 1,000', transform=ax.transAxes, ha='right', fontsize=12, color='gray')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
