# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

For advantages this option avoids clutter and allows for a direct weekly comparison side-by-side. The disadvantages include it being difficult to read between the two due to differing y-axes scales and the poor use of space such as the placement of the y-axis title.  

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

For advantages this graph gives a direct visual comparison for each day of the week with a shared axis. It is space efficient and clearly demonstrates trends when comparing the two weeks. Disadvantages of this graph include there being no legend nor any visual sequential structure to show that these lines represent data chronologically.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

From page 200.

```python

import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

fig, ax = plt.subplots(figsize=(8, 5))

# Data points
years = [2008, 2008]
prices = [400, 360]
labels = ['A', 'B']

# Plot points
ax.scatter(years, prices, color='#2c4a6e', s=80, zorder=5)

# Add labels to the left of each point
for label, x, y in zip(labels, years, prices):
    ax.text(x - 0.05, y, label, ha='right', va='center', fontsize=10, color='#333333')

# Axis formatting
ax.set_xlim(2007.5, 2014.5)
ax.set_ylim(0, 520)
ax.set_xticks(range(2008, 2015))
ax.set_yticks([0, 100, 200, 300, 400, 500])
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${int(x):,}'))

# Labels and title
ax.set_xlabel('Year', fontsize=11)
ax.set_ylabel('Average price', fontsize=11)
ax.set_title('Retail price over time', fontsize=13, loc='left', pad=10)

# Style
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(left=False)
ax.yaxis.grid(True, color='#dddddd', linewidth=0.8)
ax.set_axisbelow(True)

plt.show()
```
